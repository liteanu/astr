```python
%matplotlib inline
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
ghj = fits.open('./Downloads/radial.fits')
data = ghj[1].data
header = ghj[1].header
ghj.close()
```

```python
data
```

```python
header
```

```python
plt.style.use('default')
plt.style.use('ggplot')
```

```python
plt.scatter(data['u']-data['g'],data['g'], c = data['r'], alpha = .1)
cb = plt.colorbar()
cb.set_label('g')
plt.gca().invert_yaxis()
plt.xlabel('u-g')
plt.ylabel('g')
plt.title('CM Diagram: u-g vs g');
```

```python
plt.scatter(data['r']-data['i'],data['r'], c = data['r'], alpha = .1)
cb = plt.colorbar()
cb.set_label('r')
plt.gca().invert_yaxis()
plt.xlabel('r-i')
plt.ylabel('r')
plt.title('CM Diagram: r-i vs r');
```

```python
plt.scatter(data['r']-data['i'],data['u']-data['g'], c = data['r'], alpha = .1)
cb = plt.colorbar()
cb.set_label('r')
plt.gca().invert_yaxis()
plt.xlabel('r-i')
plt.ylabel('u-g')
plt.title('CC Diagram r-i vs u-g');
```

```python
plt.scatter(data['r']-data['z'],data['g']-data['i'], c = data['r'], alpha = .1)
cb = plt.colorbar()
cb.set_label('r')
plt.gca().invert_yaxis()
plt.xlabel('r-z')
plt.ylabel('g-i')
plt.title('CC Diagram: r-z vs g-i');
```
