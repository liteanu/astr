```python
%matplotlib inline
import numpy as np
from photutils.aperture import CircularAperture 
from photutils import aperture_photometry
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
from astropy.utils.data import get_pkg_data_filename
from matplotlib.colors import LogNorm
```

```python
hdul = fits.open('./quadRU.fits')
data = hdul[0].data
hdul.close()
```

```python
positions = [(460.296, 537.212), (298.493, 505.783), (569.258,668.795), (746.084,641.166),(685.3,600.413),(802.724,478.154),(848.312,659.125),(685.5,599.5),(895,572),(460,538)]
apertures = CircularAperture(positions, r=3.)
```

```python
phot_table = aperture_photometry(data, apertures)
print(phot_table)
```

```python
image_file = get_pkg_data_filename('./quadRU.fits')
```

```python
image_data = fits.getdata(image_file, ext=0)
```

```python
plt.figure()
plt.imshow(image_data, cmap='rainbow', norm=LogNorm())
plt.colorbar()
```
